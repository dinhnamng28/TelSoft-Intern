```python
import requests
import pandas as pd

url = "https://api.coingecko.com/api/v3/coins/markets"
params = {
    "vs_currency": "usd",                # bắt buộc
    "order": "market_cap_desc",          # sắp xếp theo vốn hóa giảm dần
    "per_page": 50,                      # lấy 50 coin
    "page": 1,                           # trang đầu tiên
    "sparkline": False                   # không cần dữ liệu sparkline
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)
    print(df[["id", "symbol", "name", "current_price", "market_cap_rank"]])
else:
    print("Lỗi khi gọi API:", response.status_code)
```

```python
df.to_csv("top_50_coin.csv", index=False, encoding='utf-8-sig')
```
